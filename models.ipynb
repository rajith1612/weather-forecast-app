```python
from flask_sqlalchemy import SQLAlchemy
from datetime import datetime

db = SQLAlchemy()

class WeatherQuery(db.Model):
    id = db.Column(db.Integer, primary_key=True)
    user_input = db.Column(db.String(255), nullable=False)  # what user typed
    normalized_name = db.Column(db.String(255), nullable=True)  # city, country
    lat = db.Column(db.Float, nullable=True)
    lon = db.Column(db.Float, nullable=True)
    start_date = db.Column(db.Date, nullable=True)
    end_date = db.Column(db.Date, nullable=True)
    created_at = db.Column(db.DateTime, default=datetime.utcnow)

    current_weather_json = db.Column(db.Text, nullable=True)  # raw API response
    forecast_json = db.Column(db.Text, nullable=True)
```
